```python
#most ordered item(sorted data according to quantity ordered)
d=data.groupby('item_name')
d=d.sum()
d=d.sort_values(['quantity'],ascending=False)
d.head(5)

```

```python
#most ordered item according to choice_description(sorted data according to quantity ordered)
d=data.groupby('choice_description')
d=d.sum()
d=d.sort_values(['quantity'],ascending=False)
d.head(5)
```
